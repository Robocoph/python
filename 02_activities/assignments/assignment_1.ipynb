### Part 1: Building the base Anagram Checker

Given two valid strings, check to see if they are anagrams of each other. If it is, return `True`, else `False`. For this part, we can assume that case letters are the same as if it was a lowercase character.

Examples of anagrams:
* Silent and Listen
* Night and Think

Example outputs:
```python
anagram_checker("Silent", "listen") # True
anagram_checker("Silent", "Night") # False
anagram_checker("night", "Thing") # True
```

````python

# In this example we don't care about the case sensitivity, hence we will convert all the text to lower casefold/caseless matching

# Approach: I am thinking about folding, sorting the string, and then comparing strings. If these are really anagrams:
# Once sorted they should ideally be the same string representation

# We should also handle for string length difference because in that situation, those are not really anagrams.

def anagram_checker(word_a, word_b):
    word_a_fold = word_a.casefold()
    word_b_fold = word_b.casefold()

    # Check lengths first because if not, these are not anagrams
    if len(word_a_fold) != len(word_b_fold):
        print(f'Sorry! The words "{word_a}" and "{word_b}" are NOT anagrams — their lengths differ.')
        return False
    # Compare sorted versions
    if sorted(word_a_fold) == sorted(word_b_fold):
        print(f'Great job! The words "{word_a}" and "{word_b}" are anagrams.')
        return True
    else:
        print(f'Sorry! The words "{word_a}" and "{word_b}" are NOT anagrams.')
        return False

anagram_checker("Silent", "listen")
````

### Part 2: Expanding the functionality of the Anagram Checker

Using your existing and functional anagram checker, let's add a boolean option called `is_case_sensitive`, which will return `True` or `False` based on if the two compared words are anagrams and if we are checking for case sensitivity.

````python
# This is the updated function for case sensitive anagrams. Here my approach changed in two places. 

# 1. I changed the variable naming in the function to word_a_str or word_a_string to denote which string I am comparing against
# 2. I incorporated is_case_sensitive first code block to ensure whether I will approach it with the fold case or leave it as is for case sensitive

def anagram_checker(word_a, word_b, is_case_sensitive):
    # code block 1: check whether we are transforming the text to  or not
    if is_case_sensitive : 
        word_a_str = word_a
        word_b_str = word_b
    else:
        word_a_str = word_a.casefold()
        word_b_str = word_b.casefold()
    # code block 2: Check for anagram
    # Check lengths first because if not, these are not anagrams
    if len(word_a_str) != len(word_b_str):
        return False
    # Compare sorted versions
    if sorted(word_a_str) == sorted(word_b_str):
        return True
    else:
        return False

# Run your code to check using the words below:
anagram_checker("Silent", "listen", False) # True
````

````python
anagram_checker("Silent", "Listen", True) # False
````
